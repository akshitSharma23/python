```python
def player_1(position):
    print("to quit the game press any key other then 'Y','y' ")
    m=input('player 1 \ninput "Y" to roll th dice')
    if m=='Y' or m=='y':
        d=dice()
        print('dice rolled by player 1')
        print(f"dice number is {d}")
        position+=d
        position=snake_ladder(position)
        if position<101:
            print(f"player 1 at {position}")
            return position
        else:
            return position
    else:
        return -1
```

```python
def player_2(position_):
    d=dice()
    print('dice rolled by player 2')
    print(f"dice number is {d}")
    position_+=d
    position_=snake_ladder(position_)
    if position_<101:
        print(f"player 2 at {position_}")
        return position_
    else:
        return position_
```
