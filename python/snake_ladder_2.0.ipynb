```python
def play():
    while True:
        n=input("enter a number of players where 1<n<11 ")
        if n.isdigit():
            n=int(n)
            if n in range(2,11):
                break
        else:
            print("please enter a integer number ")
            continue
    return n
```

```python
class player():
    def __init__(i,position):
        i.position=position
    def player_1(i,position,k):        
        m=input(f'player {k+1} turn\ninput "Y" to roll the dice ')
        if m=='Y' or m=='y':
            d=dice()
            print(f'dice rolled by player {k+1}')
            print(f"dice number is {d}")
            i.position+=d
            i.position=snake_ladder(i.position)
            if i.position<101:
                print(f"player {k+1} at {i.position} position")
                return i.position
            else:
                return i.position
        else:
            return -1
```

```python
from IPython.display import clear_output
def game():
    position=0
    exit=0
    p=[]
    n=play()
    for i in range(0,n):
        l=player(position)
        p.append(l)
    b=0
    while True:
        print("to quit the game press any key other than 'Y','y' ")
        for i in range(n):
            position=p[i].player_1(position,i)
            #position=current
            if position>100:
                    exit+=1
                    print(f"palyer {i+1} has win the game")
                    break
            if position==-1:
                    exit+=1
                    print(f"player {i+1} end the game")
                    break
            else:
                    print("\n")
                    pass
        print("\n")
        b+=1
        if exit!=0:
            print("game is finished")
            break
```
